```python
###Drop the row with index 1
df1 = df.drop(1)
df1.head()
```

```python
###Drop the column called “Embarked”
df2 = df1.drop(columns = 'Embarked')
df2.head()
```

```python
###Drop rows with null values
df2[df2.isnull().any(axis=1)]
df3 = df2.dropna(how='any',axis=0)
df3.head()
```

```python
#df3['Fare'].plot(kind='hist')
df3.hist(column='Fare', bins=50)
plt.show()
```

```python
df3.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); # ; avoid having the matplotlib verbose informations
```

```python
df3_corr = df3.corr()['Fare'][:-1] # -1 because the latest row is SalePrice
golden_features_list = df3_corr[abs(df3_corr) > 0.3].sort_values(ascending=False)
print("There is {} slightly correlated values with SalePrice:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
for i in range(0, len(df3.columns), 5):
    sns.pairplot(data=df3,
                x_vars=df3.columns[i:i+5],
                y_vars=['Fare'])
```
